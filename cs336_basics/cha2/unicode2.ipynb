a. why utf8 not utf 16 or 32

```python
string="hello everyone,我是李明"
utf8=string.encode('utf8')
utf16=string.encode('utf16')
utf32=string.encode('utf32')
print(f'utf8:{utf8}\nutf16:{utf16}\nutf32:{utf32}')
```

b.incorrect code example

```python
def decode_utf8_bytes_to_str_wrong(bytestring:bytes):
    return "".join([bytes([b]).decode("utf-8") for b in bytestring])

decode_utf8_bytes_to_str_wrong("hello".encode("utf-8"))
```
